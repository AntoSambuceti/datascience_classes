### Numéricas

#### Analisis univariados

```python
data[numerical_columns].describe()
```

```python
colors = ['salmon','yellow','blue']
fig, axes = plt.subplots(len(numerical_columns),1,figsize=(8,14))

for i, column in enumerate(numerical_columns):
    
    sns.boxenplot(data=data, x=column, color=colors[i],ax=axes[i])   
    axes[i].set_title(f'Distribución de la variable {column}')
```

```python
colors = ['salmon','yellow','blue']
fig, axes = plt.subplots(1, len(numerical_columns),figsize=(16,5))

for i, column in enumerate(numerical_columns):
   
    axes[i].hist(data[column], color=colors[i], bins=20, density=True)   
    axes[i].set_title(f'Distribución de la variable {column}')
```

#### Analisis bivariados

Podemos observar la presencia de outliers, sobre todo en la variable Income.

¿Como cambia la distribución de las variables numéricas para las diferentes categorías de la variable objetivo?
¿Estas variables ayudan a dividir la población a predecir? ¿Contribuyen a explicar mejor su comportamiento?

```python
colors = ['salmon','yellow','blue']
fig, axes = plt.subplots(1, len(numerical_columns),figsize=(16,5))

for i, column in enumerate(numerical_columns):
   
    axes[i].hist(data[data['loan_status'] == 'Approved'][column], color='green', bins=20, density=True)  
    axes[i].hist(data[data['loan_status'] == 'Denied'][column], color=colors[i], bins=20, density=True)  
    axes[i].set_title(f'Distribución de la variable {column}')
```

Observamos como estas tres variables logran dividar las poblaciones de personas con prestamos aprobados y rechazados.

```python
fig = plt.figure(figsize=(8,8))
sns.scatterplot(data=data, x='age', y='income', hue='loan_status', legend="auto")  
plt.title(f'Relación entre las variables Income, Age y Loan Status')
plt.show()
```

Las personas de edades mas bajas posen menores ingresos y a la vez son a los se le deniegan los prestamos.

```python
corr = data[numerical_columns].corr()
corr
```

### Categóricas

A las variables categóricas las podemos analizar en función a su frecuencia, también en conjunto con alguna variable numérica, o bien con la variable objetivo. 

Se suele analizar la cardinalidad de las variables categoricas, es decir número de categorías o valores distintos que puede tomar la variable. 

```python
fig = plt.figure(figsize=(22,10))
sns.countplot(data=data, x='occupation',
                color='blue')
plt.xticks(rotation=45)
plt.title("Frecuencia de Ocupaciones")
plt.ticklabel_format(style='plain', axis='y')
```

```python
data.occupation.nunique()
```

La variables Occupation posee una alta cardinalidad, y esto puede ocasionar problemas en el entrenamiento de modelos.

#### Analisis binavariado

```python
fig = plt.figure(figsize=(6,6))
sns.barplot(data=data,
                x='gender', y='age',
                estimator=np.mean)
plt.title("Promedio de edades según el género")
plt.xticks(rotation=45)
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
sns.barplot(data=data,
                x='gender', y='age',
                hue='loan_status',
                estimator=np.mean)
plt.title("Promedio de edades según el género y variable loan_status")
plt.xticks(rotation=45)
plt.show()
```

El genero de la persona que toma el prestamo no parece ser determinante en el target.

```python
fig = plt.figure(figsize=(6,6))
sns.barplot(data=data,
                y='age', x='loan_status',
                estimator=np.mean)
plt.title("Promedio de edades según target")
plt.xticks(rotation=45)
plt.show()
```

```python
data.education_level.value_counts()
```

```python
fig = plt.figure(figsize=(8,6))
sns.countplot(data=data, x='education_level',
                hue='loan_status')
plt.xticks(rotation=45)
plt.title("Promedio de edades según la educación y variable loan_status")
plt.ticklabel_format(style='plain', axis='y')
```

La variable education_level parece tener cierta incidencia en la target, encontrando cateogiras como Master's y Doctoral que poseen todos sus prestamos aprobados. 

```python
fig = plt.figure(figsize=(6,6))
sns.barplot(data=data,
                y='age', x='education_level',
                estimator=np.mean)
plt.title("Promedio de edades según education_level")
plt.xticks(rotation=45)
plt.show()
```
